```python
import numpy as np
import matplotlib.pyplot as plt
from os import mkdir
from pathlib import Path
from helper_functions import load_field

IN = Path('source_data')
FIELDS = [87, 96, 99, 103, 74, 87, 96.2, 151, 176]
```

```python
T, nu, R = load_field(103, IN)

nu_insulator = nu[np.where((nu < 1.02) & (nu > 0.98))]

for n in nu_insulator:
    idx_n = np.where(n == nu)[0][0]
    rho = R[:, idx_n]

    figure = plt.gcf()
    figure.figsize = (5, 5)
    figure.dpi = 150

    # plt.title(n)
    # plt.plot(T, rho, marker='o', markersize = 3, markerfacecolor = 'none', markeredgecolor = 'navy', linewidth = 1.0, color = 'blue')

    # plt.show()

```

```python
nu_focus =  1.001

idx_n = np.where(abs(nu_focus - nu) < 0.001)[0][0]
rho_focus = R[:, idx_n]
plt.plot(T, rho_focus)
plt.ylim(0, 1000)

plt.xlim(2, 4)
plt.show()

dpdt = np.gradient(rho_focus, T)
plt.plot(T, dpdt)

horizontal_line = np.arange(2, 4, 100)
plt.plot(horizontal_line)
plt.xlim(2, 4)
plt.ylim(0, 100)
plt.show()
```
